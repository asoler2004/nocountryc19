Se trabajarán los **outliers** con el método **IQR**


El IQR es una medida de la dispersión del 50% de los datos. Se calcula restando el primer cuartil a partir del tercer cuartil. El primer cuartil es la mediana de la mitad inferior de los datos y el tercer cuartil es la mediana de la mitad superior de los datos. El IQR representa el rango de valores que contienen la mayoría de los datos y es menos sensible a los valores atípicos que la media o la desviación estándar.

```python
# Crear una lista para almacenar los nombres de las columnas con valores binarios
binary_features = [col for col in df.columns if df[col].dropna().isin([0, 1]).all()]
```

Vemos que las variables `number_outpatient`, `number_emergency` y `number_inpatient` tienen valores inexplicables desde el pto de vista de la desviación estandard.

Una forma de identificar los valores atípicos con IQR es utilizar el valor 1.5*Regla IQR. Esta regla define los valores atípicos como valores que están por debajo de Q1 - 1.5*IQR o superior Q3 + 1.5*IQR. Estos valores se consideran demasiado alejados de la tendencia central de los datos y pueden marcarse o eliminarse. El factor 1,5 es una opción común, pero se puede ajustar en función de la distribución y el contexto de los datos.

```python
diccionario = {}
for col in non_binary_non_ids_features:
  percentile25 = df[col].quantile(0.25)
  percentile75 = df[col].quantile(0.75)
  IQR  = percentile75 - percentile25
  upper_limit = percentile75 + 1.5 * IQR
  lower_limit = percentile25 - 1.5 * IQR
  diccionario['upper_limit'+ '_' + col] = upper_limit
  diccionario['lower_limit'+ '_' + col] = lower_limit
```

```python
diccionario
```

```python
for col in non_binary_non_ids_features:
    df[col] = np.where(
        df[col] > diccionario['upper_limit_' + col],
        diccionario['upper_limit_' + col],
        np.where(
            df[col] < diccionario['lower_limit_' + col],
            diccionario['lower_limit_' + col],
            df[col]
       )
   )
```

```python
# Verificando
fig = plt.figure(figsize=[24, 16])
fig.suptitle('Boxplot de data corregida', fontsize=18, fontweight = 'bold')
fig.subplots_adjust(top = 0.92)
fig.subplots_adjust(hspace = 0.4, wspace = 0.15)

for i, col in enumerate(list(non_binary_non_ids_features)):
    ax1 = fig.add_subplot(4, 3, i+1)
    ax1 = sns.boxplot(data = df[non_binary_non_ids_features], x=col, color= colours[i])
    ax1.set_title(f'{col}', fontdict=axtitle_dict)
    ax1.set_xlabel(f'{col}', fontdict=axlab_dict)
```

En este caso se utiliza la tecnica `ADASYN (Adaptive Synthetic Sampling)`. A diferencia de otras técnicas de sobre-muestreo, como SMOTE (Synthetic Minority Over-sampling Technique), ADASYN se adapta de manera adaptativa a la distribución de la clase minoritaria. Concretamente, este algoritmo funciona con los siguientes pasos:
1. Se calcula la diferencia entre el número de instancias de la clase mayoritaria y la clase minoritaria (llamada diferencia de desequilibrio).

2. Para cada instancia de la clase minoritaria, calcula la distancia entre esta instancia y sus k vecinos más cercanos de la clase mayoritaria.

3. Calcula los pesos (o proporciones) de muestra para cada instancia de la clase minoritaria en función de la diferencia de desequilibrio y las distancias a sus vecinos.

4. Genera instancias sintéticas para las instancias de la clase minoritaria utilizando los pesos de muestra calculados. Esta "generación de instancias sintéticas" se realiza mediante la interpolación de características entre una instancia de la clase minoritaria y k sus vecinos seleccionados.

Basicamente,  ADASYN genera instancias sintéticas cerca de las regiones de la clase minoritaria que son más difíciles de clasificar, lo que ayuda a mejorar la capacidad del modelo para capturar los patrones y características de la clase minoritaria.

```python
from imblearn.over_sampling import ADASYN

adasyn = ADASYN(n_neighbors=10, sampling_strategy=.5, random_state = 123)
X_lr_resampled_adasyn, y_lr_resampled_adasyn = adasyn.fit_resample(X_lr_train, y_lr_train)
```

## Balanceo SMOTE

En este caso utilizaremos la tecnica `SMOTE (Synthetic Minority Over-sampling Technique)`. Es una técnica de sobre-muestreo u oversampling, que consite en generar nuevas muestras sintéticas para la clase minoritaria a partir de las observaciones existentes mediante la interpolación lineal entre los vecinos más cercanos de la misma clase.

Funciona tomando una muestra de la clase minoritaria para buscar sus vecinos más cercanos. Luego, se elige de forma random uno de estos vecinos y se crea un ejemplo sintético en algún lugar entre el ejemplo original y el vecino elegido. De esta manera, se aumenta el tamaño de la clase minoritaria al crear nuevas muestras que son una combinación de las muestras existentes

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy = .5)
X_lr_resampled_smote, y_lr_resampled_smote = smote.fit_resample(X_lr_train, y_lr_train)
```

## Sin balanceo de clases

En este caso haremos una validación cruzada con el método Stratified K Fold que consiste en sseleccionar n fraines diferentes de dataset tanto para los valores de train como para los valores de test teninedo en cuenta que no se repitan las muestras incluidas en cada uno

```python
from sklearn.model_selection import StratifiedKFold

model_lr_no_rebalance = LogisticRegression(max_iter=10000)
n_folds = 5
acum_accuracy=0
i=0

kf = StratifiedKFold(n_splits=n_folds)

for train_index, test_index in kf.split(X_lr, y_lr):
  i += 1

  X_train, X_test = X_lr.loc[train_index], X_lr.loc[test_index]
  y_train, y_test = y_lr.loc[train_index], y_lr.loc[test_index]

  model_lr_no_rebalance.fit(X_train, y_train)

  accuracy = model_lr_no_rebalance.score(X_test, y_test)

  acum_accuracy += accuracy
  accuracy_redondeada = round(accuracy, 5)

  print(f"Accuracy fold n° {i}: {accuracy_redondeada}")

acum_accuracy_redondeada = round(acum_accuracy/n_folds, 5)
print(f"Accuracy promedio: {acum_accuracy_redondeada}")
```

Esta prueba nos muestra que SIN HACER rebalanceo de clases, los resultados son mejores

La curva ROC (`Receiver Operating Characteristic`) se utiliza para evaluar y visualizar el rendimiento de un modelo de clasificación binaria. Proporciona una representación gráfica de la relación entre la tasa de verdaderos positivos (`True Positive Rate`) y la tasa de falsos positivos (`False Positive Rate`) a medida que se ajusta el umbral de clasificación.

El eje X de la curva ROC representa el `FPR`, que es la proporción de instancias negativas incorrectamente clasificadas como positivas. El eje Y representa el `TPR`, que es la proporción de instancias positivas correctamente clasificadas como positivas. A medida que se varía el umbral de clasificación, se obtienen diferentes pares de `TPR` y `FPR`, lo que da lugar a puntos en la curva ROC.

Idealmente, un modelo de clasificación tendría un `TPR` alto y un `FPR` bajo, lo que se traduce en una curva ROC que se acerca al vértice superior izquierdo del gráfico. Un área mayor bajo la curva ROC (AUC-ROC) indica un mejor rendimiento del modelo, donde un AUC-ROC **=** 1.0 representa una clasificación perfecta y un AUC-ROC **<=** 0.5 representa una clasificación aleatoria.

En resumen, la curva ROC permite visualizar el rendimiento de un modelo de clasificación, ayudando a seleccionar el umbral óptimo para la clasificación y a evaluar la precisión y el equilibrio entre verdaderos positivos y falsos positivos del modelo.
